# Inverted Single Pendulum

```python
#The goal of this code is to implement the inverted pendulum for the single pendulum.
#This code will utilize the dynamics of the pendulum and invert the pendulum using
#a control law derived from a Lyapunov function to change the pendulum's energy to
#match the energy of the desired angle/angular velocity and a control law derived from
#a LQR to balance the pendulum at the top.

#The pendulum system used is a simple pendulum with damping. The vertical axis going down
#from the origin represents theta=0 with counter-clockwise direction being positive.

import numpy as np
import matplotlib.pyplot as plt
import control
import matplotlib.animation as animation
```

```python
#Create a class for the pendulum simulation
class SwingUpAndBalancePendulum():

    def __init__(self,control=True,kalman=False):
        #Set control to True to see a controlled pendulum, set control to False to see a uncontrolled pendulum
        self.control=control  
        #Set kalman to True to see a pendulum with gaussian noise used with an extended Kalman Filter. 
        #set kalman to False to see a pendulum with no gaussian noise and no kalman filter.
        self.kalman=kalman 
        self.PendulumParameters()      
        self.computeLQR()
        
    #Initialize the parameters of the pendulum system
    def PendulumParameters(self,th_init=0,om_init=0):        
        self.L=1                  #Initialize the link arm length
        self.m=1                  #Initialize mass at end of the link
        self.theta_init=th_init   #Initialize the angle of the joint (in radians)
        self.omega_init=om_init   #Initialize the angular velocity of the joint (in radians/sec)
        self.g=9.81               #Initialize the acceleration due to gravity
        self.b=0                  #Initialize b (the damping coefficient)

        #Define the desired energy state (the energy for the desired angle and angular velocity)
        self.th_des=np.pi   #The desired angle
        self.om_des=0       #The desired angular velocity
        self.E_des=self.energy(self.th_des,self.om_des)    #The energy state based on desired angle and angular velocity

        #Only have gaussian noise on measurements
        if self.kalman:
            self.R_kal=np.diag((0.03,0.03)) #Measurement noise covariance matrix
            self.Q_kal=np.diag((0,0))
        
        self.u_lim=1.5     #Define the constraint on the input (limit the magnitude of the input u)
        self.k_Lyap=0.5    #Define the gain constant for the control input based on Lyapunov function
        self.dt=0.025      #Sampling time for simulation

    #Find the energy of the pendulum system
    def energy(self,theta,omega):
        T=(1/2)*self.m*self.L**2*omega**2         #Find the kinetic energy of the system
        U=-self.m*self.g*self.L*np.cos(theta)     #Find the potential energy of the system
        E=T+U                      #Find the total energy of the system
        return E
    
    #Linearize the pendulum around given theta and omega
    def lin_sys(self,theta,omega):
        A=np.array([[0,1],[-self.g/self.L*np.cos(theta),-self.b/(self.m*self.L**2)]])
        B=np.array([[0],[1/(self.m*self.L**2)]])
        return A,B
    
    #Set A and B as the matrices for the linearized pendulum system around the region of theta=np.pi and omega=0.
    #LQR is applied to obtain the gain constants to control the pendulum to stay up
    def computeLQR(self):
#         self.A=np.array([[0,1],[self.g/self.L,-self.b/(self.m*self.L**2)]])
#         self.B=np.array([[0],[1/(self.m*self.L**2)]])
        self.A,self.B=self.lin_sys(self.th_des,self.om_des)
        self.Q=np.diag((1,1))
        self.R=[2]
#         self.S=np.diag((10,1))
        self.k_LQR,self.S,e=control.lqr(self.A,self.B,self.Q,self.R)
    
    #Find the input value to drive the pendulum system
    def find_input(self,theta,omega):
        if self.control:
            #When theta and omega are close to 0 (i.e. initial theta and omega are all 0), set the input u to the input limit
            #This is done because the input u calculated using a Lyapunov function depends only on omega and not theta.
            #At the state of theta=0 and omega=0, the calculated u will be 0.
            if theta%(2*np.pi)==0 and abs(omega)==0:
                u=self.u_lim
            else:
                #Compute the error of the current states from the desired states
                xbar=np.array([[((theta)%(2*np.pi)-self.th_des)],[omega-self.om_des]])
                
                #If the cost is less than the specified value, compute input u using LQR
                if np.dot(np.transpose(xbar),np.dot(self.S,xbar))<2:
                    u=-np.dot(self.k_LQR,xbar)
                    
                #Compute the input u using the control law obtained from the Lyapunov function
                else:
                    u=self.b*omega-self.k_Lyap*omega*(self.energy(theta,omega)-self.E_des)
                    
            #If the magnitude of u is greater than the input limit, set u as the input limit        
            if abs(u)>self.u_lim:
                u=np.sign(u)*self.u_lim
        else:
            u=0      #Set input as 0 when the system is not controlled
        return u

    #Find the angular acceleration of the joint 
    def acc_func(self,theta,omega,u):
        acc=(-self.b*(omega)-self.m*self.g*self.L*np.sin(theta)+u)/(self.m*(self.L**2))
        return acc

    #The fourth order Runge-Kutta method is used to find the theta and omega of the next time step
    def RK4(self,theta,omega,u,dt):
        d_th1=dt*omega
        d_om1=dt*self.acc_func(theta,omega,u)
        d_th2=dt*(omega+d_om1/2)
        d_om2=dt*self.acc_func(theta+d_th1/2,omega+d_om1/2,u)
        d_th3=dt*(omega+d_om2/2)
        d_om3=dt*self.acc_func(theta+d_th2/2,omega+d_om2/2,u)
        d_th4=dt*(omega+d_om3)
        d_om4=dt*self.acc_func(theta+d_th3,omega+d_om3,u)

        d_th=(d_th1+2*(d_th2+d_th3)+d_th4)/6
        d_om=(d_om1+2*(d_om2+d_om3)+d_om4)/6

        theta_new=theta+d_th
        omega_new=omega+d_om

        return theta_new,omega_new
    
    #The fourth order Runge-Kutta method is used to find the covariance matrix used in the EKF 
    #and the states for the next time step
    def RK4_kalman(self,theta,omega,u,cov_mat):
        def cov_dot(th,om,cov):
            F,G=self.lin_sys(th,om)
            return np.matmul(F,cov)+np.matmul(cov,np.transpose(F))+self.Q_kal
        
        d_cov1=self.dt*cov_dot(theta,omega,cov_mat)
        th_2,om_2=self.RK4(theta,omega,u,self.dt/2)
        d_cov2=self.dt*cov_dot(th_2,om_2,cov_mat+d_cov1/2)
        d_cov3=self.dt*cov_dot(th_2,om_2,cov_mat+d_cov2/2)
        
        #This calculates the predicted states for the next time step
        th_4,om_4=self.RK4(theta,omega,u,self.dt)
        d_cov4=self.dt*cov_dot(th_4,om_4,cov_mat+d_cov3)

        #This calculates the predicted covariance matrix for the next time step
        d_cov=(d_cov1+2*(d_cov2+d_cov3)+d_cov4)/6
        cov_mat_new=cov_mat+d_cov

        return th_4,om_4,cov_mat_new
    
    #EKF update
    def kalman_update(self,th_meas,om_meas,th_pre,om_pre,cov_mat):
        #Measurement matrix
        H=np.array([[1,0],
                   [0,1]])
        
        Kalman_gain=np.matmul(np.matmul(cov_mat,np.transpose(H)),np.linalg.inv\
                              (np.matmul(H,np.matmul(cov_mat,np.transpose(H)))+self.R_kal))
        x_predicted=np.array([[th_pre],[om_pre]])
        x_measured=np.array([[th_meas],[om_meas]])
        x_update=x_predicted+np.matmul(Kalman_gain,x_measured-np.matmul(H,x_predicted))
        cov_mat_update=np.matmul((np.eye(2)-np.matmul(Kalman_gain,H)),cov_mat)
        return x_update[0,0],x_update[1,0],cov_mat_update
    
    #Simulate the pendulum system
    def simulate(self,n):
        self.th_sim=np.zeros((n))       #Initialize an array for the angles
        self.om_sim=np.zeros((n))       #Initialize an array for the angular velocities
        self.u_sim=np.zeros((n))        #Initialize an array for the control input
        self.E_sim=np.zeros((n))
        
        self.th_sim[0]=self.theta_init     #Set the initial theta to the array
        self.om_sim[0]=self.omega_init     #Set the initial omega to the array
        self.E_sim[0]=self.energy(self.th_sim[0],self.om_sim[0])       #Calculate the energy           
        
                              
        if self.kalman:
            #Initialize the arrays for estimated values
            self.th_est=np.zeros((n))
            self.th_est[0]=0             #Say that the system won't know the initial estimate and set to 0
            self.om_est=np.zeros((n))
            self.om_est[0]=0             #Say that the system won't know the initial estimate and set to 0
            self.E_est=np.zeros((n))
            self.E_est[0]=self.E_sim[0]
            
            #Initialize the arrays for measured values (with measurement noise)
            th_noise=np.random.normal(0,self.R_kal[0,0]**(0.5),n)
            om_noise=np.random.normal(0,self.R_kal[1,1]**(0.5),n)
            self.th_meas=np.zeros((n))
            self.th_meas[0]=self.theta_init+th_noise[0]
            self.om_meas=np.zeros((n))
            self.om_meas[0]=self.omega_init+om_noise[0]
            self.E_meas=np.zeros((n))
            self.E_meas[0]=self.energy(self.th_meas[0],self.om_meas[0])
            
            #Initialize the array for the covariance estimate
            self.cov_matrix=np.zeros((2,2,n))
            self.cov_matrix[:,:,0]=np.diag((1,1))
            
        #Find the theta and omega of the next time step by computing the control input and using RK4
        for x in range(1,n):
            #When there is no measurement noise, use the actual state values to calculate the input
            if not self.kalman:
                self.u_sim[x-1]=self.find_input(self.th_sim[x-1],self.om_sim[x-1])
            #When there is measurement noise, use the estimated state values to calculate the input
            else:
                self.u_sim[x-1]=self.find_input(self.th_est[x-1],self.om_est[x-1])
                
            #Find the actual states based on the input     
            self.th_sim[x],self.om_sim[x]=self.RK4(self.th_sim[x-1],self.om_sim[x-1],self.u_sim[x-1],self.dt)
            self.E_sim[x]=self.energy(self.th_sim[x],self.om_sim[x])
            
            #When there is measurement noise, calculate estimated states using Extended Kalman Filter
            if self.kalman:
                #Store the measured values with noise
                self.th_meas[x],self.om_meas[x]=self.th_sim[x]+th_noise[x],self.om_sim[x]+om_noise[x]
                self.E_meas[x]=self.energy(self.th_meas[x],self.om_meas[x])
                
                #Predict the states and covariance matrix
                th_pre,om_pre,cov_mat_pre=self.RK4_kalman(self.th_est[x-1],self.om_est[x-1],self.u_sim[x-1],self.cov_matrix[:,:,x-1])
                
                #Estimate the states and covariance matrix for the next time step
                self.th_est[x],self.om_est[x],self.cov_matrix[:,:,x]=\
                        self.kalman_update(self.th_meas[x],self.om_meas[x],th_pre,om_pre,cov_mat_pre)
                self.E_est[x]=self.energy(self.th_est[x],self.om_est[x])


        #Base (Joint 1) location at origin  j1=np.array([0,0])      
        j2=np.array([self.L*np.sin(self.th_sim[:]),-self.L*np.cos(self.th_sim[:])])  #End effector location
        return j2

    #Print the parameters
    def printParams(self):
        print("L={0:2.3f} m, m={1:2.3f} kg, g={2:2.3f} m/s^2, b={0:2.3f}".format(self.L,self.m,self.g,self.b))
        print("Theta_init={0:2.3f} rad, Omega_init={1:2.3f} rad/s".format(self.theta_init,self.omega_init))
        print("Theta_des={0:2.3f} rad, Omega_des={1:2.3f} rad/s".format(self.th_des, self.om_des))
        print("E_des={0:2.3f}, K_Lyap={1:2.3f}, dt={2:2.3f} sec".format(self.E_des, self.k_Lyap,self.dt))

    #Plot figures for the phase plot, control input over time, angle over time, angular velocity over time, and energy over time
    def plotFigures(self):
        t=np.linspace(0,pend.dt*n,n)    #Create the time array

        #Show the phase plot of the angle and angular velocity
        plt.figure(0)
        plt.plot(self.th_sim,self.om_sim,'k-')
        plt.title("Phase plot of the angle and angular velocity")
        plt.xlabel("Angle (rad)")
        plt.ylabel("Angular velocity (rad/sec)")
        plt.grid()

        #Show the plot of the control input over time
        plt.figure(1)
        plt.plot(t,self.u_sim)
        plt.title("Control input over time")
        plt.xlabel("Time (sec)")
        plt.ylabel("Control Input")
        plt.grid()

        #Show the angle of the pendulum over time
        plt.figure(2)
        plt.plot(t,self.th_sim,'r-',lw=2.5)
        plt.title("Angle over time")
        plt.xlabel("Time (sec)")
        plt.ylabel("Angle (rad)")
        plt.grid()

        #Show the angular velocity of the pendulum over time
        plt.figure(3)
        plt.plot(t,self.om_sim,'r-',lw=2.5)
        plt.title("Angular velocity over time")
        plt.xlabel("Time (sec)")
        plt.ylabel("Angular velocity (rad/s)")
        plt.grid()

        #Show the total energy of the pendulum over time
        plt.figure(4)
        plt.plot(t,self.E_sim,'r-',lw=2.5)
        plt.title("Total Energy of the pendulum system over time")
        plt.xlabel("Time (sec)")
        plt.ylabel("Energy")
        plt.grid()
        
        if self.kalman:
            #Also show the measured and estimated angles of the pendulum over time
            plt.figure(2)
            plt.plot(t,self.th_meas,'g-',lw=0.6)
            plt.plot(t,self.th_est,'b--',lw=2.5)
            plt.legend(['Actual','Measured','Estimated'])

            #Also show the measured and estimated angular velocities of the pendulum over time
            plt.figure(3)
            plt.plot(t,self.om_meas,'g-',lw=0.6)
            plt.plot(t,self.om_est,'b--',lw=2.5)
            plt.legend(['Actual','Measured','Estimated'])

            #Show the total energy of the pendulum over time
            plt.figure(4)
            plt.plot(t,self.E_meas,'g-',lw=0.6)
            plt.plot(t,self.E_est,'b--',lw=2.5)
            plt.legend(['Actual','Measured','Estimated'])
```

```python
#Set the matplotlib animation to show 1 figure during the animation
%matplotlib nbagg     

n=500     #Number of samples

#Initialize the class and simulate the pendulum
pend=SwingUpAndBalancePendulum()
j2=pend.simulate(n)

#Define the figure for the animation
fig = plt.figure(figsize=(8.333, 6), dpi=80)
ax = fig.add_subplot(111,autoscale_on=False, xlim=(-2, 2), ylim=(-2, 2))   
ax.set_aspect('equal')
line, = ax.plot([], [], 'o-', lw=2)
time_template = 'time = %.1fs'
time_text = ax.text(pend.dt, 0.9, '', transform=ax.transAxes)
    
def init():
    line.set_data([], [])
    time_text.set_text('')
    return line, time_text
    
def animate(i):
    thisx = [0, j2[0,i]]
    thisy = [0, j2[1,i]]

    line.set_data(thisx, thisy)
    time_text.set_text(time_template % (i*pend.dt))
    return line, time_text

#Show the animation
ani = animation.FuncAnimation(fig, animate, range(0, n), interval=pend.dt*1000, blit=True, init_func=init)
plt.show()

#Print the parameters
pend.printParams()
```

```python
#Set the matplotlib plots to show the plots properly
%matplotlib inline

pend.plotFigures()
```
